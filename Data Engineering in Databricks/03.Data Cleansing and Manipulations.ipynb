## Pre-requisites

### Loading the dataset

```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
                    inferSchema=True,
                    sep="|",
                    header=True,
                    quote="'",
                )
```

### Drop missing Records

```python
df.na.drop().display()
```

```python
df.na.drop(subset=["name"]).display()
```

```python
df.na.drop(how="all").display()
```

```python
from pyspark.sql.functions import col
df.filter((col('id').isNotNull()) & (col('name').isNotNull())).display()
```

### Filling the null values with 0 for integer values

```python
df.na.fill(0).display()
df.na.fill(0).na.fill("Null in source").display()
```

```python
df.na.fill({
    "id": -1,
    "gen": "NA",
    "exp":0,
}).display()
```

```python
from pyspark.sql.functions import avg
mean = df.select(avg("exp")).collect()[0][0]
df.na.fill({
    "id": -1,
    "gen": "NA",
    "exp": int(mean),
}).display()
```

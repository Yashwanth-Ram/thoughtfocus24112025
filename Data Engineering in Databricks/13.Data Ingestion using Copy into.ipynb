```python
%sql
DROP TABLE IF EXISTS quickstart_catalog.quickstart_schema.users_int;
create table quickstart_catalog.quickstart_schema.users_int(
  id int,
  name string,
  dob date,
  email string,
  gender string,
  country string,
  region string,
  city string,
  asset int,
  marital_status string
) using delta;
describe extended quickstart_catalog.quickstart_schema.users_int;
```

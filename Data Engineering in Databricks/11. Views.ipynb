## Pre-requisites

```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv", sep=",", header=True, inferSchema=True, quote="'")
```

### Views

#### Temporary views

The view is only visible or accessible within the notebook
There is no namespace for temp views

```python
df.createTempView("test_user_view")
```

```python
for view in spark.catalog.listTables():
    print(view)
```

```python
%sql

SELECT * FROM test_user_view;
```

#### Global View

The views can also be accessed from all the other notebooks

```python
df.createGlobalTempView("test_global_user_view")
```

```python
for view in spark.catalog.listTables("global_temp"):
    print(view)
```

```python
%sql

SELECT * FROM global_view.test_user_view;
```

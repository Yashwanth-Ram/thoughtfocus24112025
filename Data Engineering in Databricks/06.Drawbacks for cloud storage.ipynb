## Loading the dataset

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
df.display()
```

### Writing into parquet format

```python
df.write.format("parquet").save("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet", mode="overwrite")

df_parquet = spark.read.parquet("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet")
df_parquet.display()
```

```python
df_parquet.createOrReplaceTempView("users_vw")
```

```python
%sql

UPDATE users_vw
SET country = 'Bharat'
WHERE country = 'India'
```

### Disadvantage: Overwriting existing directory and no roll back

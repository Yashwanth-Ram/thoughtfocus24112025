## Pre-requisites

### Load employee.csv into dataframe

```python
emp_df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv", header=True, inferSchema=True, sep="|", quote="'")
display(emp_df)
```

## APPly python function inside pyspark

```python
def convert_to_title(st):
  if st is None:
    return None
  else:
    return st.title()
```

```python
emp_df.select("id","name").display()
```

```python
emp_df.select("id",convert_to_title("name")).display()
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
convert_to_title_udf = udf(convert_to_title, StringType())
# convert_to_title_udf = udf(lambda e:convert_to_title(e) if e is not None else None)
```

```python
emp_df.select("id",convert_to_title_udf("name")).display()
```

## Applying the UDF inside SQL

```python
emp_df.createOrReplaceTempView("employee_vw")
```

```python
spark.udf.register("convert_to_title_udf_sql", convert_to_title, StringType())
```

```python
%sql

SELECT id, convert_to_title_udf_sql(name) from employee_vw LIMIT 10
```
